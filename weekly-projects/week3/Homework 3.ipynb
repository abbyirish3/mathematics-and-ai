```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os, time
```

## Part 1: Data preparation

```python
# toggle settings
add_noise = True

# Initiliaze lists for image collection
train_images = []
test_images = []

for i, images in enumerate([train_images, test_images]):

    # set paths to images of pandas and bears in train and test set
    datasetname = ['Train','Test'][i]
    folder_path1 = 'PandasBears/{}/Pandas/'.format(datasetname)
    folder_path2 = 'PandasBears/{}/Bears/'.format(datasetname)
    

    for folder_path in [folder_path1, folder_path2]:

        # print the name of the folder that is currently being processed
        print(folder_path, end=' ')
        
        # go through all files in the folder
        file_count = 0
        for filename in os.listdir(folder_path):
            
            # find the files that are JPEGs
            if filename.endswith(".jpg") or filename.endswith(".jpeg"):

                # add 1 to the file count
                file_count += 1
                
                # Construct full file path
                file_path = os.path.join(folder_path, filename)
    
                # import image
                image = plt.imread(file_path, format='jpeg')
    
                # convert to gray scale
                image = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])
    
                # decrease image size by 50%
                image = image[::2,::2]

                if add_noise:
                    # add some noise
                    image = image + np.random.normal(scale=100, size=image.shape)
    
                # add the new image to collection
                images.append(image)

        print('has {} images'.format(file_count))
```

```python
# look at 4 random bears
for i0, i in enumerate(np.random.randint(0, 500, size=4)):
    plt.subplot(2,2,1+i0)
    plt.imshow(train_images[i][::2,::2],cmap='Greys_r')
```

# Part 2: Singular value decomposition

```python
# 1. Follow the steps in the eigenfaces tutorial from the UW databook [[link]](https://github.com/dynamicslab/databook_python/blob/master/CH01/CH01_SEC06_2_3_4.ipynb) to perform an SVD on the images.
# 2. Note that you will need to "center the data" before doing the SVD. 
# Data centering means replacing each variable $X_i$ with a new variable $X_i'$ that is equal to $X_i$ minus $X_i$'s mean value. (Think carefully about whether you want to use the mean of train set, test set, or the full data set for this.)
# 3. Display the first four "eigenbears" (i.e., the images associated with the first four eigenvectors). Explain what you see.
# 4. The singular vectors are sometimes called "principal components". Principal component analysis (PCA) is a data analysis method for which one projects high-dimensional data into the subspace of the first two singular vectors. 
# Use the code scaffold provided below do perform PCA for the bears data. What do you notice?

# # Construct data matrix of centered data
train_mat = []
test_mat = []

for image in train_images: 
    flat_image_train = np.ndarray.flatten(image)
    train_mat.append(flat_image_train)
train_mat = np.column_stack(train_mat)

for image in test_images:
    flat_image_test = np.ndarray.flatten(image)
    test_mat.append(flat_image_test)
test_mat = np.column_stack(test_mat)


mean_image = np.mean(train_mat, axis = 1, keepdims = True) # goes through each row to take the average 

# center the data 
train_mat = train_mat - mean_image
test_mat = test_mat - mean_image

print(np.shape(train_mat))
print(np.shape(test_mat))

# Perform SVD with centered data matrix
U, S, Vt = np.linalg.svd(train_mat)
```

```python
# inspect shapes of matrices
print(f'u.shape:{U.shape},s.shape:{S.shape},v.shape{Vt}')
print(np.shape(U))

# show the first four eigenbears
for i in range(4):
    plt.subplot(2,2,1+i)
    plt.imshow((U[:,i]).reshape((128,128)), cmap='Greys_r',
        # force colormap to be the same for all four
        vmin=-np.max(np.abs(U[:,:4])),
        vmax=np.max(np.abs(U[:,:4])))
    plt.colorbar()
plt.subplots_adjust(wspace=0.4)
```

The bears are very blurry here, but we can see a rough outline of their facial features.

```python
# Clustering of panda bears and brown bears along the first and second principal component

# indices of pandas in the test set
indices_pandas = range(50)
# indices of brown bears in the test set
indices_brownbears = range(50,100)

for i, indices in enumerate([indices_pandas, indices_brownbears]):
    # get projections of data onto principal component 1
    p1 = [np.dot(U[:,0], np.ravel(test_images[x])) for x in indices]
    # get projections of data onto principal component 2
    p2 = [np.dot(U[:,1], np.ravel(test_images[x])) for x in indices]
    plt.plot(p1, p2, marker='+x'[i], lw=0, label=['Pandas', 'Grizzlies'][i])

# annotate axes
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')
# add legend
plt.legend()
```

# Part 3: Nonparametric classification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# construct response variable: Train set was created by appending 250 pandas
# and THEN 250 brown bears to the list of training images. We code pandas as
# '0' and brown bears as '1'.
y_train =  np.concatenate([np.zeros(250), np.ones(250)])

# Test set was created by appending 50 pandas and THEN 50 brown bears to the 
# list of test images. We code pandas as '0' and brown bears as '1'.
y_test = np.concatenate([np.zeros(50), np.ones(50)])
```

```python
print('   k\t|  # errors\t| misclassified bears')
print('--------------------------------------------')

all_errors = []

for k in range(1,16):
    # fit KNN model
    modelKN = KNeighborsClassifier(n_neighbors=k).fit(train_mat.T, y_train)
    # use model to make predictions on the test set
    predictions = [modelKN.predict([np.ravel(test_images[i])]) for i in range(len(y_test))]
    # detect misclassifications
    errors = np.abs((np.array(predictions).T)[0]-y_test)
    # print results to table
    print('    {}\t|      {} \t| {}'.format(k, int(np.sum(errors)), (np.argwhere(errors).T)[0]))

    for error in np.argwhere(errors).T[0]:
        if error not in all_errors:
            all_errors.append(error)

print(all_errors)
```

The most difficult bears to classify (without noise) seem to be 5, 21, 22, 32, 39, 41, and 42.
The most difficult bears to classify (with noise) seem to be 5, 6, 8, 21, 22, 24, 39, 41, 42, 9, 32, and 36.

```python
# Show hard-to-classify bears

print(all_errors) # list where each object is a smaller list 

num_cols = 4
num_rows = int(np.ceil(len(all_errors) / num_cols))

for i, error in enumerate(all_errors):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(U[:,i].reshape(128, 128), 
                cmap='Greys_r', 
                vmin = -np.max(np.abs(U[:,:4])), 
                vmax = np.max(np.abs(U[:,:4]))) # force colormap to be the same for all four
    plt.colorbar()
    plt.title(f"Bear #{error}", fontsize = 10) # add title indicating which bear it is

plt.subplots_adjust(wspace = .75, hspace = .5) # room inbetween and above plots
plt.show()
        
```

# Part 4: Parametric classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
# 1. Try using logistic regression and LDA to classify the bears. 
 
# logistic regression: 
log_model = LogisticRegression().fit(train_mat.T, y_train) # fit model using training data 

log_predictions = log_model.predict(test_mat.T) # use model to make predictions on the test set

err = 0
for i in range(len(log_predictions)):
    if log_predictions[i] != y_train[i]:
        err += 1

print("percent errors:", err/len(y_test), "%")
# log_errors = np.abs(log_predictions - y_test) # detect misclassifications

# list_log_errors = np.argwhere(log_errors).flatten().tolist() # make list of errors

# print(list_log_errors)


# LDA:
lda_model = LinearDiscriminantAnalysis().fit(train_mat.T, y_train) # fit model using training data

lda_predictions = lda_model.predict(test_mat.T) # use model to make predictions on the test set

lda_errors = np.abs(lda_predictions - y_test) # detect misclassifications

list_lda_errors = np.argwhere(lda_errors).flatten().tolist() # make list of errors

print(list_lda_errors)
```

2. What method gives you the best test accuracy? 
Logistic regression gives the best test accuracy because there are no misclassfied bears.

3. How does the result compare to the non-parametric classification?
Both logistic regression and LDA seem to have better test accuracy than non-parametric classification, which had much more errors. 

```python
#4 of Part 4: Bear mask

# Construct the bear mask:
# Show absolute value of logistic-regression coefficients for each pixel
plt.imshow(np.abs((log_model.coef_).reshape((128,128))))
plt.colorbar()
```

The pixels with the highest coefficients are shown in lime green and yellow and mostly occur towards the top left and top right of the image, presumably where the ears of the bear would be. This makes sense given that pandas have black ears while brown bears do not, and thus this could be an identifying feature for the model in classifying the image. There is also some green around (40, 40) and (80, 40), presumably where the eyes of the bear would be. This again makes sense because pandas have dark black fur around their eyes while brown bears do not, and thus this area again could be an important idenitfying feature for the model.
